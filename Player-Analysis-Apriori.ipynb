```python
def countContributions(arr):
    countContribution={}
    for val in arr:
        for vals in val :
            if(vals not in countContribution):
                countContribution[vals]=0
            countContribution[vals]+=1
            
        
    return countContribution        
```

```python
countContribution = countContributions(Goals)
print("Goal Contribution of Each Player")
countContribution
```

```python
def support_count(countContribution, noOfGoals):
    support=[]
    for vals in countContribution:
        val=round((countContribution[vals]/noOfGoals)*100, 2)
        support.append(val)

    return support    
```

```python
support=support_count(countContribution, len(Goals))
```

```python
supportDict={'Player':list(countContribution.keys()), 'Support':support}
playerContributionSupport=pd.DataFrame(supportDict, columns=['Player', 'Support'])
```

```python
def support_threshold(contribution, support, min_support):
    playerSupportThreshold=[]
    for val in range(len(support)):
        if(support[val]>=min_support):
            playerSupportThreshold.append(contribution[val])
            
    return playerSupportThreshold         
```

```python
playerSupportThreshold=support_threshold(list(countContribution.keys()), support, min_support)
print("Players with atleast", min_support,"% Contributions")
maxContrib={}
for val in range(len(playerSupportThreshold)):
    maxContrib[playerSupportThreshold[val]]=round(support[list(countContribution.keys()).index(playerSupportThreshold[val])])
    
maxContrib=pd.DataFrame.from_dict(maxContrib, orient='index')
maxContrib
```

```python
playerPartnership=list(permutations(playerSupportThreshold, 2))
```

```python
def combo_find(playerPartnership, Goals):    
    combinePlayers = []
    iteration = 0
    for player in playerPartnership:
        for g in Goals:
            iteration = 0
            ids=[0,0]
            i=0
            for p in player:
                if p in g:
                    iteration+=1
                    ids[i]=(g.index(p))
                    i+=1
            if (ids[1]==ids[0]+1 and iteration == len(player)):
                combinePlayers.append(player)
   
    return(combinePlayers)
```

```python
combinePlayers=combo_find(playerPartnership, Goals)
```

```python
def counting_combo(combinePlayers):
    count_combo={}
    for combo in range (len(combinePlayers)):
        if(combinePlayers[combo] not in count_combo):
            count_combo[combinePlayers[combo]]=0
        count_combo[combinePlayers[combo]]+=1
    
    return(count_combo)    
```

```python
count_combo=counting_combo(combinePlayers)
c1=list(count_combo.keys())
c2=list(count_combo.values())
```

```python
def unpacking_tuple(arr, arr2, arr3):
    unpack=[]
    
    for i in arr:
        f, s = i
        if f not in unpack:
            unpack.append(f)
        if s not in unpack:
            unpack.append(s)
        
            
    ind=[]
    for val in range(len(unpack)):
        if(unpack[val] in arr2):
            ind1=arr2.index(unpack[val])
            ind.append(arr3[ind1]) 
    

    return(unpack,ind)
```

```python
def totals(arr1, dict1, arr2, arr3):
    tot=[]
    for val in range(len(arr1)):
        if(arr1[val] in dict1):
            key=arr2.index(arr1[val])
            tot.append(arr3[key])
    return(tot)           
```

```python
def confidence(supportDict, playerPairs):
    conf=[]
    
    for i in range(len(playerPairs.keys())):
        comb=list(playerPairs.keys())[i]
        player=list(playerPairs.keys())[i][0]
        
        suppUnion=list(playerPairs.values())[i]
        suppPlayer=supportDict['Support'][supportDict['Player'].index(player)]
        
        confs=(suppUnion/suppPlayer)*100
        
        conf.append(confs)
    
    finalPasses={'Players':pairThresh, 'Confidence':conf}
    final=pd.DataFrame(finalPasses, columns=['Players', 'Confidence'])
    
    return final
```
